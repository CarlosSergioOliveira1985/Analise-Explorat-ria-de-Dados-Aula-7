2 - Transformações dos dados envolvendo junção de dados, transformações de atributos e/ou normalização

Transformando renomeando e normalizando os dados para junção de tabelas Produção de tonelada de Cevada por Hectar com Rendimento na produção de Cereais tonelada por hectar

```python
BF_USA = pd.merge(colunas_selecionadas_USA,Colunas_fertilizer_USA,on=['País','Ano'])
BF_USA.head()
```

4 Uma análise de correlação

Verificando Graficamente a Corelação entre

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(12, 6))

# Supondo que você já tenha os dataframes BF_USA com as colunas 'Ano', 'Feijão Ton. Hec.' e 'Rend. Cer. Hec'
sns.lineplot(x='Ano', y='Cevada Ton. Hec.', data=BF_USA, label='Cevada')
sns.lineplot(x='Ano', y='Rend. Cer. Hec.', data=BF_USA, label='Fertilizantes')

plt.title('Produção de Cevada vs. Red. Cereais com Fertilizantes nos EUA', fontsize=14)
plt.legend()
plt.xticks(rotation=90)
plt.xlabel('Ano')
plt.ylabel('Toneladas por Hectare')
plt.show()
```

```python
tractors = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-09-01/cereal_yields_vs_tractor_inputs_in_agriculture.csv')
```

```python
tractors.head()
```

```python
# Filtrando os dados dos tratores dos Estados Unidos
tractors_USA = tractors[tractors.Entity == 'United States']

# Criando uma nova coluna com os valores ajustados
Colunas_tractors_USA = tractors_USA[['Entity', 'Year', 'Tractors per 100 sq km arable land']]
Colunas_tractors_USA['Tractors per 100 sq km arable land (ajustado)'] = Colunas_tractors_USA['Tractors per 100 sq km arable land'] / 100

# Exibindo as primeiras linhas do DataFrame com a nova coluna
Colunas_tractors_USA.head()
```

```python
Colunas_tractors_USA = Colunas_tractors_USA[Colunas_tractors_USA.Entity == 'United States']

# Criando um novo DataFrame apenas com as colunas relevantes
Colunas_tractors_USA = Colunas_tractors_USA[['Entity', 'Year', 'Tractors per 100 sq km arable land (ajustado)']]

# Exibindo as primeiras linhas do DataFrame resultante
Colunas_tractors_USA.head()
```

```python
novos_nomes = {
    'Entity': 'País',
    'Year': 'Ano',
    'Tractors per 100 sq km arable land (ajustado)': 'Cento Tratores para 100 km2'
}

Colunas_tractors_USA.rename(columns=novos_nomes, inplace=True)
Colunas_tractors_USA.head()
```

```python
Colunas_tractors_USA.dtypes
```

```python
Colunas_tractors_USA['Ano'] = Colunas_tractors_USA['Ano'].astype('int64')
Colunas_tractors_USA.head()
```

5 - Uma análise de informação mútua, emprego de atributos derivados ou teste de interdependência

```python
BFT_USA = pd.merge(BF_USA, Colunas_tractors_USA, on=['País', 'Ano'])

BFT_USA.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supondo que você já tenha o dataframe BFT_USA com as colunas 'Cevada Ton. Hec.', 'Rend. Cer. Hec.' e 'Cento Tratores para 100 km2'
BFT_USA_cor = BFT_USA[['Cevada Ton. Hec.', 'Rend. Cer. Hec.', 'Cento Tratores para 100 km2']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(BFT_USA_cor, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlação')
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(12, 6))

# Supondo que você já tenha os dataframes BF_USA com as colunas 'Ano', 'Feijão Ton. Hec.' e 'Rend. Cer. Hec'
sns.lineplot(x='Ano', y='Cevada Ton. Hec.', data=BFT_USA, label='Feijão')
sns.lineplot(x='Ano', y='Rend. Cer. Hec.', data=BFT_USA, label='Fertilizantes')
sns.lineplot(x='Ano', y='Cento Tratores para 100 km2', data=BFT_USA, label='Trator')

plt.title('Produção de Feijão vs. Red. Cereais com Fertilizantes nos EUA', fontsize=14)
plt.legend()
plt.xticks(rotation=90)
plt.xlabel('Ano')
plt.ylabel('Toneladas por Hectare')
plt.show()

```
